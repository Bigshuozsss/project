```python
# 拆分时间列
full['date']=full['datetime'].map(lambda s:s.split(' ')[0])
full.head()

full['year']=full['date'].map(lambda s:s.split('-')[0])
full.head()
```

```python
#拆分月份
import calendar
full['month']=full['date'].map(lambda s:s.split('-')[1])
full['month']=full['month'].map(lambda s:calendar.month_name[int(s)])
full.head()
```

```python
full['hour']=full['datetime'].map(lambda s:s.split(' ')[1].split(':')[0])
full.head()
```

```python
from datetime import datetime
full['weekday']=full['date'].map(lambda s:datetime.strptime(s,'%Y-%m-%d').weekday())
full['weekday']=full['weekday'].map(lambda s:calendar.day_name[s])
full.head()
```

```python
# 特征工程
weekdayDf=pd.DataFrame()
weekdayDf=pd.get_dummies(full['weekday'],prefix='weekday')
weekdayDf.head()
```

```python
monthDf=pd.DataFrame()
monthDf=pd.get_dummies(full['month'],prefix='month')
monthDf.head()
```

```python
full.drop(['datetime','atemp','season','workingday','date','month','weekday','year'],axis=1,inplace=True)
full=pd.concat([full,weekdayDf,monthDf],axis=1)
full.info()
```

```python
full=full.convert_objects(convert_numeric=True)
full.info()
```

```python
# 随机森林预测
myIndex=test['datetime']
myIndex.head(3)
```

```python
from math import log,exp 
def mylog(data):  
    data=int(data)+1  
    return log(data)
```

```python
labelData=full['casual'].loc[0:10885].apply(func=mylog).values
labelData2=full['registered'].loc[0:10885].apply(func=mylog).values
```

```python
trainData=full.drop(['casual','registered','count'],axis=1).loc[0:10885].values
testData=full.drop(['casual','registered','count'],axis=1).loc[10886:].values
```

```python
# casual
from sklearn.ensemble import RandomForestRegressor  
rfModel=RandomForestRegressor(n_estimators=100)  
rfModel.fit(trainData,labelData)  
preds=rfModel.predict(testData).tolist()
preds=[exp(i)-1 for i in preds]
```

```python
# registered
rfModel=RandomForestRegressor(n_estimators=100)  
rfModel.fit(trainData,labelData2)  
preds2=rfModel.predict(testData).tolist()
preds2=[exp(i)-1 for i in preds2]
```

```python
countList=np.sum([preds,preds2],axis=0).tolist()
countList=[round(i,0) for i in countList]
predDf=pd.DataFrame({'count':countList,'datetime':myIndex})
predDf.head()
predDf.info()
```

```python
predDf['count']=predDf['count'].astype(int)
predDf.info()
```

```python
datetime=predDf.datetime
predDf=predDf.drop('datetime',axis=1) 
predDf.insert(0,'datetime',datetime) #把datetime这一列放到第一列
predDf.head()
predDf.to_csv('bike_pred.csv',index=False)
```
